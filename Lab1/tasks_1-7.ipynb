```python
# 1. Загрузить данные с Kaggle – выбрать датасет (например, Титаник).

from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
df = pd.read_csv("DataSets/train.csv")  # путь к файлу
```

```python
# 2. Вывести с помощью python данные из датасета на экран.

# настройки вывода
pd.set_option('display.max_columns', None)   # показывать все столбцы

print("Информация о датасете:\n")
df.info()

print("\nПЕРВЫЕ 5 СТРОК:\n")
display(df.head())

print("\nПОСЛЕДНИЕ 5 СТРОК:\n")
display(df.tail())

print("\nПЕРВЫЕ 105 СТРОК:\n")
display(df.head(105))
```

```python
# 3. Получить количество пропущенных значений для каждого столбца в датасетах.

print("Количество пропущенных значений в каждом столбце:\n")
print(df.isnull().sum())
```

```python
# 4. Заполнить пропущенные значения в датасете модой/медианой/средним значением и показать, что они действительно были заполнены.

# 1) Заполнение числовых столбцов медианой или средним значением.
# Выберем только числовые колонки
numeric_data_df = df.select_dtypes(include='number')

# Заполним медианой (устойчиво к выбросам)
for column in numeric_data_df.columns:
    median_value = df[column].median()
    df[column].fillna(median_value, inplace=True)

# 2) Заполнение категориальных столбцов модой.
# Выберем нечисловые колонки
categorical_data_df = df.select_dtypes(exclude='number')

for column in categorical_data_df.columns:
    mode_value = df[column].mode()[0]   # самое частое значение
    df[column].fillna(mode_value, inplace=True)

print("\nКоличество пропущенных значений ПОСЛЕ заполнения:\n")
print(df.isnull().sum())

print("\nПроверим несколько строк, чтобы убедиться, что значения заменены:\n")
display(df.head(105))

print("Также выведем информацию о датасете, для проверки количества ненулевых значений в столбцах:\n")
df.info()
```

```python
# 5. Провести нормализацию данных.

# Выделим числовые данные (с тем же порядком и индексом)
numeric_df = df.select_dtypes(include='number')
numeric_columns = numeric_df.columns.tolist()
print("Числовые данные, к которым будем применять нормализацию:\n", numeric_columns)

# 1) Минимально-максимальное масштабирование
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(numeric_df)

# Преобразуем результат обратно в DataFrame, а именно создаем новый DataFrame с тем же индексом, что и исходный df, чтобы потом безопасно вставить их в df для пункта 6.
minmax_scaled_df = pd.DataFrame(minmax_scaled, columns=numeric_columns, index=numeric_df.index)
print("\nПример нормализованных данных (MinMaxScaler, значения в диапазоне [0, 1]):")
display(minmax_scaled_df.head())

# Запишем преобразованные данные (через MinMaxScaler()) обратно в основной df
df[numeric_columns] = minmax_scaled_df.astype(float)  # float чтобы не конфликтовало с Int64/NA

# 2) Нормализация по Z-оценке
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(numeric_df)

standard_scaled_df = pd.DataFrame(standard_scaled, columns=numeric_df.columns)
print("\nПример нормализованных данных (StandardScaler):")
display(standard_scaled_df.head())

# 3) Проверим статистику
print("\nПроверим статистику после StandardScaler (среднее mean ~ 0, а стандартное отклонение std ~ 1):")
display(standard_scaled_df.describe().loc[['mean', 'std']])
```

```python
# 6. Провести преобразование категориальных данных так, чтобы в будущем это не привело к переобучению модели машинного обучения и датасет имел удобоваримый вид.

# 1) Уберём идентификаторы и колонки, бесполезные для модели
df_proc = df.copy()
cols_to_drop = ['PassengerId', 'Name']  # ID/Name — уникальны, ведут к переобучению
for c in cols_to_drop:
    if c in df_proc.columns:
        df_proc.drop(columns=c, inplace=True)

# 2) Разберём Cabin на Deck / Num / Side (если колонка есть)
if 'Cabin' in df_proc.columns:
    cabin_split = df_proc['Cabin'].str.split('/', expand=True)
    cabin_split.columns = ['CabinDeck', 'CabinNum', 'CabinSide']
    df_proc = pd.concat([df_proc.drop(columns=['Cabin']), cabin_split], axis=1)

# 3) Приведём логические колонки к 0/1
# (CryoSleep и VIP были object, приведём к 0/1)
for col in ['CryoSleep', 'VIP', 'Transported']:
    if col in df_proc.columns:
        # некоторые значения могут быть "False"/"True" в виде строк, или bool
        df_proc[col] = df_proc[col].map({True: 1, False: 0, 'True': 1, 'False': 0}).astype('Int64')

# 4) Список категориальных признаков, которые хотим OHE-преобразовать.
cat_cols = []
for c in df_proc.select_dtypes(exclude='number').columns:
    # пропускаем целевой (если он не в cat), а также пропускаем колонки, которые уже числовые
    nunique = df_proc[c].nunique(dropna=False)
    if nunique <= 30:
        cat_cols.append(c)
    else:
        # Для очень высококардинальных колонок (напр. ID-подобные) — можно заменить на частотное кодирование.
        print(f"ВНИМАНИЕ: столбец '{c}' имеет высокую уникальность ({nunique}) — нужно либо убрать столбец, либо рассмотреть частотное/целевое кодирование.")

print("Категориальные для OHE:", cat_cols)

# 5) One-Hot Encoding с drop_first=True (чтобы избежать мультиколлинеарности и уменьшить размерность (не будет HomePlanet_Earth, CabinDeck_A, CabinSide_P))
df_final = pd.get_dummies(df_proc, columns=cat_cols, drop_first=True)

# 6) Результат
print("\nИсходная форма:", df.shape)
print("Форма после преобразований:", df_final.shape)
display(df_final.head(15))
```
